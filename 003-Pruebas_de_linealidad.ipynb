```python
###############################################################################
import torch as torch
#from math import sqrt, exp
###############################################################################

###############################################################################
def otraFuncionNOx():
  y = 100
  return y;


###############################################################################
def sigmoid(x):
  """
  Sample sigmoid function
  param x: input array
  return y: sigmoid output
  """
  y = 1 / (1 + torch.exp(-1 * x))
  return y;

###############################################################################
def linealFunction(x):
  """
  Sample lineal function
  param x: input array
  return y: 5x
  """
  y = 5*x 
  return y

###############################################################################
def funcion2(x):
  y = 5*x + 13
  return y

###############################################################################
def funcion3(x):
  y = x.sqrt(  )
  return y

###############################################################################
def funcion4(x):
  y = x * otraFuncionNOx()
  return y

###############################################################################
def funcion5(x):
  y = x.log( )
  return y

###############################################################################
def funcion6(x):
  y = x.abs( )
  return y

###############################################################################
def funcion7(x):
    
  y = x.max( )
  print("x ",x)
  return y

###############################################################################
def funcion8(x):
  y = x + 6
  return y

###############################################################################
def checkLinearity(operator): 
  torch.manual_seed(0)
  
  # Random number 
  x = torch.randn(1)
  print("x", x )
  
  #sample scalars
  alpha = torch.randn(1);
  beta = torch.randn(1)
  print("alpha ", alpha)
  print("beta: ", beta)
 
  #sample functions
  f1 = lambda x: x + 2
  f2 = lambda x: x * x
  #############################################
  # prueba con la función sigmoide
  #f2 = lambda x: 1 / (1 + torch.exp(-1 * x))
  #############################################
 
  #compute left and right side of the equation  
  right = alpha * operator(f1(x)) + beta * operator(f2(x))
  left = operator(alpha * f1(x) + beta * f2(x))
  print(left, right)
  
  difference = torch.abs(right - left);  
  
  if difference < 0.0001:            
      mensaje = "Si es lineal " + str(difference)
  else:
      mensaje = "No es lineal " + str(difference)
  return mensaje
  

###############################################################################
# Programa principal  

print("----------------------------------------------")
prueba = "y = 5*x"
print(prueba, " --- ", checkLinearity(linealFunction))

print("----------------------------------------------")
prueba = "y = 1 / (1 + torch.exp(-1 * x))"
print(prueba, " --- ", checkLinearity(sigmoid))

print("----------------------------------------------")
prueba = "y = 5*x + 13"
print(prueba, " --- ", checkLinearity(funcion2))

print("----------------------------------------------")
prueba = "y = sqrt( x )"
print(prueba, " --- ", checkLinearity(funcion3))

print("----------------------------------------------")
prueba = "y = x * otraFuncionNOx()"
print(prueba, " --- ", checkLinearity(funcion4))

print("----------------------------------------------")
prueba = "y = x.log( )"
print(prueba, " --- ", checkLinearity(funcion5))

print("----------------------------------------------")
prueba = "y = x.abs( )"
print(prueba, " --- ", checkLinearity(funcion6))

print("----------------------------------------------")
prueba = "y = x.max( )"
print(prueba, " --- ", checkLinearity(funcion7))


print("----------------------------------------------")
prueba = "y = x + 6"
print(prueba, " --- ", checkLinearity(funcion8))
```
